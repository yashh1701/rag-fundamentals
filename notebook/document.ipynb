TextLoader

```python
# Load the text file using TextLoader
from langchain_community.document_loaders import TextLoader

loader=TextLoader("../data/text_files/python_intro.txt",encoding="utf-8")
document=loader.load()
print(document)
```

Directory Loader

```python
# Load multiple text files from a directory using DirectoryLoader 
# This is useful when you have a collection of documents you want to process together.
from langchain_community.document_loaders import DirectoryLoader

# load all the text files from the directory
dir_loader=DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",          # Pattern to match files, * means match any number of characters, ** means match any number of directories 
    loader_cls= TextLoader,   #loader class to use
    loader_kwargs={'encoding': 'utf-8'}, # Additional arguments to pass to the loader class 
    show_progress=False              # Whether to show a progress bar during loading 
)
print(dir_loader.load()) # This will load all the text files in the specified directory and return a list of Document objects.
```

```python
# Load PDF documents using PyPDFLoader or PyMuPDFLoader
# PyPDFLoader is a simple PDF loader that uses the PyPDF2 library to extract text from PDF files. 
# It is suitable for basic PDF documents but may struggle with complex layouts or scanned PDFs.
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader

## load all the pdf's from the directory
dir_loader=DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf", ## Pattern to match files  
    loader_cls= PyMuPDFLoader, ##loader class to use
    show_progress=False

)

pdf_documents=dir_loader.load()
pdf_documents
```
